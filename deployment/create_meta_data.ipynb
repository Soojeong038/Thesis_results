```python
import pandas as pd
```

```python
info = pd.read_csv("ext_info_2022.csv", index_col =0)
info.head()
```

```python
ents = pd.read_csv("doc_ent_freq_1.csv", index_col =0)
ents.head()
```

```python
info['entities'] = ""
info.head()
```

```python
doc_id = list(info['doc_id'])
doc_count = 0
for i in doc_id:
    ent_dict={}
    temp = ents[ents['doc_id'] == i]
    temp = temp.reset_index(drop = True)
    for k in range(len(temp)):
        if temp['freq'][k] > 1:
            ent_dict.update({temp['org_name'][k] : temp['freq'][k]})
    ent_dict = sorted(ent_dict.items(), key=lambda x: x[1], reverse=True)
    info.at[doc_count, 'entities']  = ent_dict
    doc_count += 1
                                         
info
```
