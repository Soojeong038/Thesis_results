```python
import spacy
trained_nlp = spacy.load("/content/drive/MyDrive/output_roberta_100/model-best")
```

```python
data['entities'] = ''

for i in range(len(data)):
  doc = trained_nlp(data['txt'][i])
  temp = []
  for ent in doc.ents:
    temp.append(ent.text)
  data.at[i,'entities'] = temp

data.head()
```
